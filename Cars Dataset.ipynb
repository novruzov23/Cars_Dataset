```python
def df(data2, head=5):
    print('####### head #########')
    print(data2.head(head))
    print('######## shape #######')
    print(data2.shape)
    print('######## ndim ########')
    print('######### describe #######')
    print(data2.describe().T)
    print('######## isnull ########')
    print(data2.isnull().sum())
    print(data2.ndim)
    print('######### tail ###########')
    print(data2.tail(head))
  
    
df(data)
```

```python
# eksik olan değerler bulunması, sıralı bir şekil de ve oranlarına bakılması
def missin_values(dataframe):
    isnull=[col for col in dataframe.columns if dataframe[col].isnull().sum() >0]
    sorts=dataframe[isnull].isnull().sum().sort_values(ascending=False)
    ratio=(dataframe[isnull].isnull().sum() / dataframe.shape[0]*100)
    concat=pd.concat([sorts, np.round(ratio)], axis=1, keys=['missing_value', 'ratio'])
    print(concat)
```

```python
missin_values(data)
```
